### Task: Understand Museum Attendance


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup Complete')
```

```python
# Load data
data_path = '../data/seabornData/museum_visitors.csv'
museum_df = pd.read_csv(data_path, index_col = 'Date', parse_dates = True)

museum_df.head()
```

The Firehosue Museum claims they ran an event in 2014 that brought a big number of visitors, and that they should get an extra budget to run a similar event this year.

```python
plt.figure(figsize=(16, 6))
plt.title("Museum Attendance Over Time")
sns.lineplot(data=museum_df)
```

### Assess Seasonality
When meeting with the employeeys of the `Avila Adobe`, you hear that the one major pain point is that the number of visitors varies greatly with the season.

Part A:
- Create a `line chart` that shows how the number of visitors to the `Avila Adobe` has evolved over time
- Replace `inf` with `NaN`, and then fill them in with 0

```python
import numpy as np

# Replace infinite values with NaN
museum_df = museum_df.replace([np.inf, -np.inf], np.nan)

# Now we can use the fillna() method to replace NaN with zeros
museum_df = museum_df.fillna(0)
```

```python
# Plot the data

plt.figure(figsize=(16, 6))
plt.title('Month Visitors to Avila Adobe')
plt.xlabel('Date')
plt.ylabel('Visitor Count')
sns.lineplot(data=museum_df['Avila Adobe'], label='Attendance')
```

## Bar Plots

Dataset: US Department of Transportation that tracks flight delays

```python
# Load dataset
file_path = '../data/seabornData/flight_delays.csv'
flight_delay_df = pd.read_csv(file_path, index_col = 'Month')

flight_delay_df.head()
```

```python
plt.figure(figsize=(16, 6))
plt.title('Average Arrival Delay for Spirit Airlines by Month')
sns.barplot(x=flight_delay_df.index, y=flight_delay_df['NK'], data = flight_delay_df)
plt.ylabel('Arrival Delay in Minutes')
```

# Heat Maps

```python
# Plot heatmap for all data
plt.figure(figsize=(16, 6))
plt.title('Avg Arrival Delay for Each Airline, by Month')
sns.heatmap(data = flight_delay_df, annot=True)
plt.xlabel('Airline')
```
