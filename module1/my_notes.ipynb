One of the most common magic commands is the `%matplotlib inline` command. This command will enable plots to be displayed directly inside the notebook.

````python
%matplotlib inline
````

````python
"""Simple demo of scattermplot"""
import numpy as np
import matplotlib.pyplot as plt

n = 50
x = np.random.rand(n) # create an array of n random numbers
y = np.random.rand(n) # create an array of n random numbers
colors = np.random.rand(n) # create an array of n random numbers
# Calculate the area
area = np.pi * (15 * np.random.rand(n))**2 # 0 to 15 point radius
# Plot data
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
````

# Conda

Create and manage your virtual environments.

To create virtual environment with a specific `python` version.

```bash
conda create -n cs458 python=3.10 #creates a new environment with name 'cs4580', and python version
```

Use `conda activate <environment name>` to activate an environment. `conda deactivate` will deactivate a current environment.

To get the list of current packages installed:
`conda list`

To install packages, use:
`conda install <name>`
`conda install numpy` (example)

## Comparison with other alternatives

#### Python list vs Numpy
- Elements of different types `vs` same type
- variable size nesting is possible `vs` same size
- No operations at the list level `vs` operations allowed at the array level
- uses a lot of memory `vs` less memory
