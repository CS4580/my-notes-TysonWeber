```python
import pandas as pd
# Load data
reviews_df = pd.read_csv('../data/wine-reviews/winemag-data-130k-v2.csv')
```

```python
reviews_df.columns
```

```python
# Task: Who are the most common wine reviewers?
# Create a series whose index is 'taster_twitter_handle'
# whose values count how many reviews the reviewer has

# Method 1: using groupby() and size()
reviews_written = reviews_df.groupby('taster_twitter_handle').size()

# Method 2: using groupby() and count(). Displays all column counts
reviews_written = reviews_df.groupby('taster_twitter_handle').count() # displays more information

# Method 3: produces the same output as Method 1
reviews_written = reviews_df.groupby('taster_twitter_handle')['taster_twitter_handle'].count()

reviews_written
```

```python
# Task: What is the best wine I can buy for a given amount of money?
# Create a series whose index is the wine price and whose values are the max number of points
# a wine costing that much was given in a review.
# Sort the series by price, ascending ($4 at the top and $3300 at the bottom)

reviews_written = reviews_df.groupby('price')[['points']].max().sort_index()
reviews_written
```

```python
# Task: What are the minimum and maximum prices for each variety of wine?
# Create a DataFrame whose index is 'variety' and whose values are min and max values

price_ends = reviews_df.groupby('variety').price.agg(['min','max']) # Use the agg() method to collect the min and max values

price_ends
```

```python
# Task: What are the most expensive wine varieties?
# Create a variable 'sorted_varieties' containing a copy of the 
# DataFrame from the previous task

sorted_varieties = price_ends.sort_values(by=['min', 'max'], ascending=False)
sorted_varieties
```

```python
# Task: Create a Series whose index is 'reviewers' and whose values is
# the average review score given by the reviewer
# Hint: need 'taster_name' and 'points'

reviews_mean_rating = reviews_df.groupby('taster_name').points.mean().sort_values(ascending=False)
reviews_mean_rating
```
