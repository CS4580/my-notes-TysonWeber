# Introduction to Plotting

## Python packages for plotting:

- `matplotlib`: All plots in pandas are matplotlib objects
- `seaborn`: Focused on statistical visualization
- `bokeh`: To create interactive visualization for the internet
- `Plotly`: Create interactive embedded

### Mathplotlib
Main submodule is `matplotlib` which will give you different functions to work with different plot elements.

### Creating a plot

- Import the necessary packages
- Import or create data
- Call the appropriate plotting functions

```python
import pandas as pd
import matplotlib.pyplot as plt
# Display the plots inside the notebook
%matplotlib inline
```

```python
# Dummy data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

print(x)
print(y)
```

```python
# Default is a line plot
plt.plot(x,y)

plt.show()
```
